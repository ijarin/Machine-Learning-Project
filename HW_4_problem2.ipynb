```
# Assign input x and observed output b
import numpy as np
import matplotlib.pyplot as plt
import math
import random
x=np.arange(0,1.1,.1)
pii=math.pi
#if initial data was noiseless
b_clean=np.sin(2*pii*x)
#now there are noisy initial data
b0=np.sin(2*pii*x)-.1
randomList = []
# Set a length of the list to 10
for i in range(0, 11):
    # any random numbers from 0 to 1000
    randomList.append((random.randint(-50, 100))/100)
randn=np.array(randomList)

#generate b
b=b0+.2*randn
```

```
#Function to create matrix
def calc_matrix(x,m):
  n=len(x)
  #m=3
  A=np.zeros((n,m+1))
  for i in range(m+1):
      A[:,i]=x**i
      #A[:,1]=x
      #A[:,2]=np.square(x)
      #A[:,3]=x**3
  
  return A
```

```
def weight_calc(A):
  #calculate w
  a1=np.dot(np.transpose(A),A)
  a1_inv=np.linalg.inv(a1)
#a1_inv.shape
  a2=np.dot(a1_inv,np.transpose(A))
  a3=np.dot(a2,b)
  w=a3
  return w
```

```
#calculate the matrix when M=0
#calculate the matrix A
A0=calc_matrix(x,0)
#Calculate weights for w, M=0
w0=weight_calc(A0)
w0
```

```
#calculate prediction for y when M=0
y0=np.dot(A0,w0)
```

```
#plot
plt.plot(x, b, linestyle='--',marker='o',color='black',label='Observation with Noise',linewidth=1.5)  
plt.plot(x, b_clean,linestyle='--', marker='*',color='blue',label='Observation Without Noise',linewidth=1.5) 
plt.plot(x, y0,linestyle='--', marker='+',color='red',label='Prediction at M=0',linewidth=1.5)
plt.xlabel('x',fontsize='15')
plt.ylabel('y',fontsize='15')
plt.legend()
```
