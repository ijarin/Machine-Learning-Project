```
# Assign input x and observed output b
import numpy as np
import matplotlib.pyplot as plt
import math
x=np.arange(0,1.1,.1)
pii=math.pi
b=np.sin(2*pii*x)

```

```
#create the third (m=3) order polynomial matrix X
def calc_matrix(x,m):
  n=len(x)
  #m=3
  A=np.zeros((n,m+1))
  for i in range(m+1):
      A[:,i]=x**i
      #A[:,1]=x
      #A[:,2]=np.square(x)
      #A[:,3]=x**3
  
  return A

```

```
#calculate the matrix A
A=calc_matrix(x,3)
A
```

```
#calculate w
a1=np.dot(np.transpose(A),A)
a1_inv=np.linalg.inv(a1)
#a1_inv.shape
a2=np.dot(a1_inv,np.transpose(A))
a3=np.dot(a2,b)
w=a3
print(w)
```

```
#calculate y when m=3
y3=np.dot(A,w)
#y3
```

```
#The taylor series expansion of sin(x)=x-(1/3!)x^3
x1=np.arange(-.5,.6,.1)
xt=2*pii*x1               #here x=xt
y_taylor=xt-(xt**3)*(1/math.factorial(3))
```

```
plt.plot(x, b, linestyle='--',marker='.',color='black',label='Actual Observation',linewidth=2)  
plt.plot(x, y3,linestyle='--', marker='.',color='blue',label='Prediction when m=3',linewidth=2) 
plt.plot(x1, y_taylor,linestyle='--', marker='.',color='red',label='Taylor Series Calculation when m=3',linewidth=1)
plt.xlabel('x',fontsize='15')
plt.ylabel('y',fontsize='15')
plt.legend()
```

```
#m=9 calculate A
A9=calc_matrix(x,9)
```

```
#calculate w
a19=np.dot(np.transpose(A9),A9)
a19_inv=np.linalg.inv(a19)
#a1_inv.shape
a29=np.dot(a19_inv,np.transpose(A9))
a39=np.dot(a29,b)
w9=a39
print(w9)
```

```
#calculate y when m=3
y9=np.dot(A9,w9)
```

```
#taylor series for m=9
xt=2*pii*x1
#y_taylor9=x**9/362880 - x**7/5040 + x**5/120 - x**3/6 + x
y_taylor9=(xt**9/math.factorial(9)) - (xt**7/math.factorial(7))+(xt**5/math.factorial(5)) - (xt**3/math.factorial(3) )+ xt
```

```
plt.plot(x, b, linestyle='--',marker='o',color='black',label='Actual Observation',linewidth=2)  
plt.plot(x, y9,linestyle='--', marker='.',color='blue',label='Prediction when m=9',linewidth=2) 
plt.plot(x1, y_taylor9,linestyle='--', marker='.',color='red',label='Taylor Series when m=9',linewidth=1)
plt.xlabel('x',fontsize='15')
plt.ylabel('y',fontsize='15')
plt.legend(loc=1,fontsize=8)
```
